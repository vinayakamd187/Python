#load and explore the dataset

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/Naresh312143/mahalingpur-python-ml/refs/heads/main/Salary_Data.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

##feature scaling(very important for SVR)

```python
from sklearn.preprocessing import StandardScaler
x=df[['Years of Experience']].values
y=df[['Salary']].values
sc_x=StandardScaler()
sc_y=StandardScaler()
x_scaled=sc_x.fit_transform(x)
y_scaled=sc_y.fit_transform(y)
```

##flatten y for SVR

```python
y_scaled
```

```python
y_scaled=y_scaled.ravel()
```

```python
y_scaled
```

##train SVR model(RBF kernel)

```python
from sklearn.svm import SVR
svr=SVR(kernel='rbf',C=100,epsilon=0.1)
svr.fit(x_scaled,y_scaled)
```

##predictions and inverse scaling

```python
y_pred_scaled=svr.predict(x_scaled)
y_pred=sc_y.inverse_transform(y_pred_scaled.reshape(-1,1))
```

```python
y_pred_scaled
```

```python
y_pred
```

##visualization:actual vs predicted

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color='blue',label='Actual')
plt.plot(x,y_pred,color='red',label='SVR prediction')
plt.title("SVR-salary predicted")
plt.xlabel("Years of Exprience")
plt.ylabel("Salary")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
x_grid=np.arange(min(x.ravel()),max(x.ravel()),0.1)
x_grid=x_grid.reshape(-1,1)
x_grid_scaled=sc_x.transform(x_grid)
y_grid_scaled=svr.predict(x_grid_scaled)
y_grid=sc_y.inverse_transform(y_grid_scaled.reshape(-1,1))
plt.scatter(x,y,color='blue',label='Actual Data')
plt.plot(x_grid,y_grid,color='red',label='SVR Smooth Curve',linewidth=2)
plt.title("SVR (RBF Kernel)-Smooth Prediction Curve")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.grid(True)
plt.legend()
plt.show()
```

##evaluate model performance

```python
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np
mse=mean_squared_error(y,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y,y_pred)
print("MSE=",mse)
print("RMSE=",rmse)
print("R2 Score=",r2)
```

##compare kernels using code

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
x=df[['Years of Experience']].values
y=df[['Salary']].values
sc_x=StandardScaler()
sc_y=StandardScaler()
x_scaled=sc_x.fit_transform(x)
y_scaled=sc_y.fit_transform(y).ravel()
kernels=['linear','poly','rbf']
for kernel in kernels:
  if kernel=='poly':
    model=SVR(kernel=kernel,degree=2)
  else:
    model=SVR(kernel=kernel)
  model.fit(x_scaled,y_scaled)
  y_pred_scaled=model.predict(x_scaled)
  y_pred=sc_y.inverse_transform(y_pred_scaled.reshape(-1,1))
  score=r2_score(y,y_pred)
  print(f"kernel:{kernel.upper()} | r2 Score:(score:4f)")
  plt.scatter(x,y,color='gray',label='Actual Data')
  plt.plot(x,y_pred,label=f'{kernel.upper()}Kernel',linewidth=2)
  plt.title(f'SVR using{kernel.upper()}Kernel')
  plt.xlabel("Years of Experience")
  plt.ylabel("Salary")
  plt.grid(True)
  plt.legend()
  plt.show()
```

Try differnt c and epsilon values

```python
C_values=[2,20,200]
epsilon_values=[0.01,0.1,1]
x_grid=np.arange(min(x.ravel()),max(x.ravel()),0.1)
x_grid=x_grid.reshape(-1,1)
x_grid_scaled=sc_x.transform(x_grid)
for c in C_values:
  for epsilon in epsilon_values:
    svr=SVR(kernel='rbf',C=c,epsilon=epsilon)
    svr.fit(x_scaled,y_scaled)
    y_grid_scaled=svr.predict(x_grid_scaled)
    y_grid=sc_y.inverse_transform(y_grid_scaled.reshape(-1,1))
    plt.figure(figsize=(8,5))
    plt.scatter(x,y,color='gray',label='Actual')
    plt.plot(x_grid,y_grid,color='red',label=f'C={c},epsilon={epsilon}')
    plt.title('SVR Salary predicted')
    plt.xlabel('Position Level')
    plt.ylabel('Salary')
    plt.legend()
    plt.grid(True)
    plt.show()
```
